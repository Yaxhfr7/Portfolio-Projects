```python
# Parsing JSON string into Python dictionary
# Saving the DataFrame to a CSV file
# Converting JSON data to a DataFrame
# Importing json to handle JSON data
# Importing pandas for data manipulation
import pandas as pd
from requests import Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
from time import sleep

df = pd.DataFrame()  # Initialize global dataframe

def api_runner():
    global df

    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

    parameters = {
        'start': '1',
        'limit': '15',   # Use a small number while testing
        'convert': 'USD'
    }

    headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': '26f3701f-bae8-4aff-87de-d7491252c453',
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        print("Status Code:", response.status_code)
        data = json.loads(response.text)

        # âœ… This part should be inside try block and NOT under except
        df2 = pd.json_normalize(data['data'])
        df2['time_stamp'] = pd.to_datetime('now')
        df = pd.concat([df, df2], ignore_index=True)
        print("Data fetched and appended to df âœ…")

    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print("Request Error:", e)

# ðŸ•’ Loop to run every 5 seconds
for i in range(333):
    api_runner()
    print(f"API run {i+1} completed")
    sleep(5)  # wait 5 seconds

# Optional: Save to CSV after all runs
df.to_csv("crypto_data.csv", index=False)
# compare it with alex file first

```
